### - ReprÃ©sentation des courbes:

```python
def funct_top(val):
    if val in range(len(Top_10_ind.object_id)):
#        session_uuid_len_1 = df.loc[(df.object_id == Top_10_ind.object_id[val])].groupby(['actor']).session_uuid.count()
        score_max_1 = df.loc[(df.object_id == Top_10_ind.object_id[val])].groupby(['actor']).score.max()
        temps_sum_1 = df.loc[(df.object_id == Top_10_ind.object_id[val])].groupby(['actor']).temps.sum()
        dictionn = {'score_max': score_max_1, 'temps_total': temps_sum_1}
        TOP_ = pd.DataFrame(dictionn)
        #TOP.temps_total = TOP.temps_total/60
        #TOP_ = TOP.sort_values(by = 'temps_total', ascending = True)
        TOP__ = TOP_.reset_index()
#        TOP_.loc[TOP_['score_max'] >= 50, 'resultat_final'] = 1
#        TOP_.loc[TOP_['score_max'] < 50, 'resultat_final'] = 0
        return TOP__
    else : 
        print('la valeur entrer est incorrect')
```

```python
Top_1 = funct_top(0)
Top_2 = funct_top(1)
Top_3 = funct_top(2)
Top_4 = funct_top(3)
Top_5 = funct_top(4)
Top_6 = funct_top(5)
Top_7 = funct_top(6)
Top_8 = funct_top(7)
Top_9 = funct_top(8)
Top_10 = funct_top(9)
```

```python
def funct_tri(n):
    var_top = funct_top(n).sort_values(by = 'temps_total', ascending = True)
    var_top.temps_total = var_top.temps_total/60
    return var_top
```

```python
funct_tri(0)
```

```python
def fig_top(n):
    plt.figure(figsize=(15, 5), layout='constrained')
    for i in range(n):
        plt.plot(funct_tri(i).temps_total, funct_tri(i).score_max, label=Top_10_ind.object_id[i])
        plt.xlabel('temps_total')
        plt.ylabel('score_max')
        plt.title("fig_top_module")
        plt.legend();
```

```python
def funct_tri_TMod(n):
    jointure_3['temps_global_min'] = temps_global
    N_modules = jointure_3.reset_index().sort_values(by = 'Taux_Rf', ascending = False)
    for i in range(n):
        df1 = pd.DataFrame(N_modules.iloc[:i+1, 0])
        df2 = pd.DataFrame(N_modules.iloc[:i+1, 4])
        df3 = pd.DataFrame(N_modules.iloc[:i+1, 5])
        df4= pd.DataFrame(N_modules.iloc[:i+1, 6])

    df = pd.concat([df1, df2, df3, df4], axis=1)
    df.insert(1, 'Rang_Module', range(1, 1 + len(df)))
    return df
```

```python
funct_tri_TMod(1)
```

```python
def funct_TM_scatterM(n):
    fig, ax = plt.subplots(figsize=(10, 5), layout = 'constrained')
        
    ax.scatter(funct_tri_TMod(n).Rang_Module, funct_tri_TMod(n).temps_global_min, s = 60)

    ax.grid(True)
    ax.set_xlabel("Rang_Module")
    ax.set_ylabel("Temps_global")
    #ax.legend();
```

```python
def funct_top_10_scatterA(n):
    fig, ax = plt.subplots(figsize=(9, 4), layout = 'constrained')
    for i in range(n):
        df = funct_top(i).sort_values(by = 'score_max', ascending = False)
        df.insert(0, 'Rang', range(1, 1 + len(df)))
        y = df.temps_total

        ax.scatter(df.Rang, y/60, s = 60, label = Top_10_ind.object_id[i])

        ax.grid(True)
        ax.set_xlabel("Rang")
        ax.set_ylabel("Temps_Total")
        ax.legend()
```

```python
def camembert():
    jointure_3.loc[jointure_3.Taux_Rf >= 80, 'Classe'] = 'Classe_A'
    jointure_3.loc[(jointure_3.Taux_Rf >= 60) & (jointure_3.Taux_Rf < 80), 'Classe'] = 'Classe_B'
    jointure_3.loc[(jointure_3.Taux_Rf >= 50) & (jointure_3.Taux_Rf < 60), 'Classe'] = 'Classe_C'
    jointure_3.loc[jointure_3.Taux_Rf < 50, 'Classe'] = 'Classe_D'
    Top_cam = jointure_3.reset_index()
    Top_cam = Top_cam.sort_values(by = 'Taux_Rf', ascending = False)

    plt.figure(figsize = (8, 8))
    x = Top_cam.groupby(['Classe']).Classe.count()
    #x.index.values
    plt.pie(x.values, labels = x.index.values,
               colors = ['red', 'green', 'yellow', 'blue'],
               explode = [0, 0.1, 0, 0],
               autopct = lambda x: str(round(x, 2)) + '%',
               pctdistance = 0.7, labeldistance = 1.1,
               shadow = False)
    plt.legend();
```

```python
camembert()
```

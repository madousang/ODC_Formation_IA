* Fonction de groupement de colonnes selon les dataset ci-dessus

```python
def regrouperSelon(dataframe, col_list,agg_dic):
    if dataframe == 'df_studentVle':
        _dataFrame = df_studentVle.groupby(col_list).agg(agg_dic)
        return _dataFrame
    elif dataframe == 'df_studentInfo':
        _dataFrame = df_studentInfo.groupby(col_list).agg(agg_dic)
        return _dataFrame
    elif dataframe == 'df_studentAssessment':
        _dataFrame = df_studentAssessment.groupby(col_list).agg(agg_dic)
        return _dataFrame
```

### - Normalisation

* Convertion de la date en jour à la date en seconde dans la colonne **date_unregistration**

```python
_dataset_['date_unregistration'] = _dataset_.date_unregistration*24*3600
```

* Utilisation du tranformeur **ColumnTransformer** qui prend en paramètre **MinMaxScaler** qui est un algorithme mettant les données sur une echelle de 0 à 1.

```python
ct_cl = ColumnTransformer([
    ("num_preprocess_MinMax", MinMaxScaler(), ["sum_click", "score", "date_unregistration", "num_of_prev_attempts", "studied_credits"])])
```

```python
Y_cl = _dataset_['final_result']
X_cl = _dataset_.drop('final_result',axis=1)
```

* Normalisation des données

```python
X_cl_ = pd.DataFrame(ct_cl.fit_transform(X_cl),
                     columns = ["sum_click", "score", "date_unregistration", "num_of_prev_attempts", "studied_credits"])
```

```python
X_cl.drop(X_cl[["sum_click", "score", "date_unregistration", "num_of_prev_attempts", "studied_credits"]], axis=1, inplace = True)
```

```python
X_cl_fl = pd.concat([X_cl, X_cl_], axis = 1)
X_cl_fl
```

#### - Separation des données en ensemble de test et d'entrainement

```python
X_train_cl, X_test_cl, y_train_cl, y_test_cl = train_test_split(X_cl_fl, Y_cl, random_state=2, test_size = .30)
```

```python
X_train_cl = X_train_cl.to_numpy()
y_train_cl = y_train_cl.to_numpy()
```

```python
X_test_cl = X_test_cl.to_numpy()
y_test_cl = y_test_cl.to_numpy()
```

### Modèle de classification descente de gradiant stockastique : SGDClassifier
* Formulation mathématique de la solution : Pk+1 = Pk − αk .grad f (Pk ).
* Où f (Pk ) est la fonction de perte à minimiser en **critère d'optimisation**.
* Comme **approche d'optimisation** on applique la descente de gradiant à cette fonction de perte grad f (Pk ), **Le gradient d'une fonction** à plusieurs variables en un certain point est un **vecteur** qui caractérise la variabilité de cette fonction au voisinage de ce point.
* Pk est le paramètre à optimiser.
* L'hyperparamètre alpha (**vitesse d'apprentissage** en anglais **learning_rate**) est aussi utiliser pour la regurarisation .

```python
clf = SGDClassifier(verbose = 0, learning_rate="optimal", random_state = 2)
```

* Creation d'un pipeline prenant en paramètre l'algorithme **OneHotEncoder** et notre modèle de classification.
* Le **OneHotEncoder** est utilisé pour encoder nos données categoricielles

```python
pipe_ = make_pipeline(OneHotEncoder(handle_unknown='ignore'), clf)
print(pipe_)
```

```python
# pipe_.get_params()
```

```python
# Entrainement du pipeline 
pipe_.fit(X_train_cl, y_train_cl)
```

* Utilisation de la **methode de grille** pour la recherche de hyperparamètre optimal alpha.

```python
parameters = {
    'sgdclassifier__alpha': np.logspace(-4, -1, 20)
}
```

```python
Mgrid = GridSearchCV(pipe_, parameters, n_jobs = -1)
```

```python
# Entrainement de l'estimateur de grille
Mgrid.fit(X_train_cl, y_train_cl)
```

```python
# Alpha optimal
Mgrid.best_params_ 
```

```python
# le meilleur score
Mgrid.best_score_
```

* Sauvegarde du modèle en local sous le nom **problème_1_class**.

```python
joblib.dump(value = Mgrid.best_estimator_, filename = 'problème_1_class.pkl')
```

```python
# L'estimateur pour le bon score et le bon paramètre
best_Mgrid = Mgrid.best_estimator_
```

### - Inférence : Généralisation du modèle avec l'utilisation des données de validations (test)

```python
# Score des données de test
best_Mgrid.score(X_test_cl, y_test_cl)
```

* Prediction sur les données de test

```python
Mgrid_predict = best_Mgrid.predict(X_test_cl)
```

* Calcul de l'accuracy : Dans la classification multiétiquette, cette fonction calcule la précision du sous-ensemble : l'ensemble d'étiquettes prédit pour un échantillon doit correspondre exactement à l'ensemble d'étiquettes correspondant dans y_true.

```python
metrics.accuracy_score(y_test_cl, Mgrid_predict)
```
